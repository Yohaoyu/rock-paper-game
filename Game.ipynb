### Set the strategies and rules

```python
import pandas as pd
```

```python
strategies=["Rock","Paper","Scissors","Spock","Lizard"]
```

```python
payoff={("Scissors","Paper"):(1,0),
        ("Paper","Rock"):(1,0),
        ("Rock","Lizard"):(1,0),
        ("Lizard","Spock"):(1,0),
        ("Spock","Scissors"):(1,0),
        ("Scissors","Lizard"):(1,0),
        ("Lizard","Paper"):(1,0),
        ("Paper","Spock"):(1,0),
        ("Spock","Rock"):(1,0),
        ("Rock","Scissors"):(1,0),
        ("Scissors","Rock"):(0,1),
        ("Rock","Spock"):(0,1),
        ("Spock","Paper"):(0,1),
        ("Paper","Lizard"):(0,1),
        ("Lizard","Scissors"):(0,1),
        ("Scissors","Spock"):(0,1),
        ("Spock","Lizard"):(0,1),
        ("Lizard","Rock"):(0,1),
        ("Rock","Paper"):(0,1),
        ("Paper","Scissors"):(0,1),
        ("Scissors","Scissors"):(0,0),
        ("Paper","Paper"):(0,0),
        ("Rock","Rock"):(0,0),
        ("Lizard","Lizard"):(0,0),
        ("Spock","Spock"):(0,0),}
```

```python
# Choose a strategy ramdomly
from random import choice
choice(strategies)
```

### Add all players

```python
names=['Jim','Jane','Peter','Zoe','Paul','Brigitte','Alex','Chris','Tom']
```

```python
# Create the society
society=[{'name':n,'score':0,'strategy':None} for n in names]
society
```

### Pair all players

```python
import itertools 
for player1,player2 in itertools.combinations(society,2):
    print(player1,player2)
```

### Play the game

```python
for player1,player2 in itertools.combinations(society,2):
    player1['strategy']=choice(strategies)
    player2['strategy']=choice(strategies)
    result=payoff[player1['strategy'],player2['strategy']]
    player1['score']+=result[0]
    player2['score']+=result[1]
```

### Show the result

```python
# Only the list
society
```

```python
# Table
socialResults=pd.DataFrame(society)
socialResults
```

```python
# Who is winner?
filter1=socialResults.score==socialResults.score.max()
winner=socialResults[filter1]
winner
```

```python
# Who is loser?
filter2=socialResults.score==socialResults.score.min()
loser=socialResults[filter2]
loser
```
